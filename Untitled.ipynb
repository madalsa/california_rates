```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('baseline_bills_Feb5.csv')

# Scenario groups
scenario_groups = {
    'Original SDGE Rates': [
        ('pre_fixed_tou_bill', 'TOU-DR'),
        ('post_fixed_tou_bill', 'TOU-DR-F')
    ],
    'Fixed Charge Progression\n(No policy changes)': [
        ('new_F0_WF0_ROE0_bill', 'F0%'),
        ('new_F25_WF0_ROE0_bill', 'F25%'),
    ],
    'Policy Changes\n(No fixed charges)': [
        ('new_F0_WF1_ROE0_bill', 'Remove WF'),
        ('new_F0_WF0_ROE0.5_bill', 'Reduce ROE'),
        ('new_F0_WF1_ROE0.5_bill', 'WF + ROE')
    ],
    'Combined: F25% + Policies': [
        ('new_F25_WF1_ROE0.5_bill', 'F25 + WF + ROE'),
    ]
}

# Check which columns exist
print("Checking columns...")
for group_name, scenarios in scenario_groups.items():
    print(f"\n{group_name}:")
    for col, label in scenarios:
        exists = col in df.columns
        print(f"  {label:20s}: {col:30s} {'✓' if exists else '✗ MISSING'}")

# Find global min and max (only existing columns)
all_cols = []
for scenarios in scenario_groups.values():
    for col, _ in scenarios:
        if col in df.columns:
            all_cols.append(col)

if len(all_cols) == 0:
    print("\nERROR: No matching columns found!")
    print("Available columns:", df.columns.tolist())
else:
    global_min = df[all_cols].min().min()
    global_max = df[all_cols].max().max()
    y_range = global_max - global_min
    y_min = max(0, global_min - 0.1 * y_range)
    y_max = global_max + 0.1 * y_range

    # Create plot (2x2)
    fig, axes = plt.subplots(2, 2, figsize=(16, 10))
    axes = axes.flatten()

    colors = ['lightgray', 'lightblue', 'lightcoral', 'lightgreen']

    for idx, (group_name, scenarios) in enumerate(scenario_groups.items()):
        ax = axes[idx]
        
        bill_data = []
        labels = []
        
        # Only add if column exists
        for col, label in scenarios:
            if col in df.columns:
                bill_data.append(df[col].values)
                labels.append(label)
        
        if len(bill_data) > 0:  # Only plot if we have data
            positions = np.arange(len(labels))
            bp = ax.boxplot(bill_data, positions=positions, widths=0.6, 
                            patch_artist=True, showfliers=False)
            
            for patch in bp['boxes']:
                patch.set_facecolor(colors[idx])
                patch.set_alpha(0.7)
            
            # Add median values
            for i, data in enumerate(bill_data):
                median = np.median(data)
                ax.text(i, median, f'${median:.0f}', 
                       ha='center', va='bottom', fontsize=9, fontweight='bold')
            
            ax.set_xticks(positions)
            ax.set_xticklabels(labels, fontsize=10)
            ax.set_ylabel('Annual Bill ($)', fontsize=11)
            ax.set_title(group_name, fontsize=11, fontweight='bold')
            ax.set_ylim(y_min, y_max)
            ax.grid(axis='y', alpha=0.3)
        else:
            ax.text(0.5, 0.5, 'No data available', 
                   ha='center', va='center', transform=ax.transAxes)

    plt.tight_layout()
    plt.savefig('all_scenarios_comparison.png', dpi=150, bbox_inches='tight')
    print("\nSaved: all_scenarios_comparison.png")

    # Print summary statistics
    print("\n" + "="*80)
    print("ANNUAL BILL SUMMARY STATISTICS")
    print("="*80)

    for group_name, scenarios in scenario_groups.items():
        print(f"\n{group_name}:")
        for col, label in scenarios:
            if col in df.columns:
                bills = df[col]
                print(f"  {label:20s}: Mean=${bills.mean():6.0f}  Median=${bills.median():6.0f}")
```
